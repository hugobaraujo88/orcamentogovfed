<font size="6">1. Criando dataframes (pandas) para alguns arquivos<font size="5">

```python
import data_prep as dp
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

despesa_exec = dp.CSVReader('dados/despesas-execucao')
orcamento_despesa = dp.CSVReader('dados/orcamento-despesa')
rceitas = dp.CSVReader('dados/receitas')

#Dataframe com os dados da execução da despesa
df_despesa_exec= despesa_exec.read_file(201401,'Despesas') #Janeiro de 2014

#Dataframe com os dados do orçamento da despesa (previsão de despesas)
df_orcamento_despesa = orcamento_despesa.read_file(2014,'OrcamentoDespesa.zip') #2014

#Dataframe com os dados das receitas arrecadadas e previstas
df_receitas = rceitas.read_file(2014,'Receitas') #2014
```

<font size="6">2. Análise da <u>Execução das Despesas</u><font size="5">

```python
df_despesa_exec.head(10)
```

É necessário escolher, dentre os valors disponíveis: 'Valor Empenhado', 'Valor Liquidado' ou 'Valor Pago', a métrica para quantidade de despesa executada.


Entre estes, ressalta-se que o 'Valor Empenhado' corresponde à primeira etapa da execução orçamentária. Trata-se de uma espécie de crédito para que se possa pagar a despesa. Neste sentido, nem todo valor empenhado converter-se-á em despesa. Por outro lado, o valor pago refere-se aos desembolsos ou saída de caixa. Essa métrica também não é a ideal, pois pode existir despesas incorridas e não pagas pela administração.


Portanto, em termos gerenciais, a métrica mais adequada para medir a despesa executada seria o Valor Liquidado.


Além da data em que a despesa fora lançada no sistema, é preciso escolher qual campo representa melhor o nome ou tipo da despesa. Para isso, é interessante veficar as colunas do dataframe: 

```python
dp.ListFormatter(df_despesa_exec.columns).print_formatted()
```

<br>
Uma análise da lista acima permite inferir que os campos 'Nome Função', 'Nome Subfunção', 'Nome Ação', 'Nome Categoria Econômica', 'Nome Grupo de Despesa', 'Nome Elemento de Despesa' e 'Modalidade da Despesa' são candidatos para representar o nome da despesa. O que se espera, por exemplo, é que a despesa esteja associada a um termo do tipo 'educação' ou 'saúde'.
<br>

```python
column_names = ['Nome Função', 'Nome Subfunção', 'Nome Ação', 'Nome Categoria Econômica','Nome Grupo de Despesa', 'Nome Elemento de Despesa', 'Modalidade da Despesa'] #campos a serem analisados

dp.DataFrameProcessor(df_despesa_exec, column_names).process_data() #Mostra, lado a lado, os valores únicos de cada coluna escolhida para a análise.
```

Claramente, o campo 'Nome Função' é o melhor para identificar as despesas. Os outros ou são muito específicos, ou são muito técnicos.

<font size="5">2.1 Conclusão:</font>

Do conjunto de dados inicial, são necessários para fins deste projeto apenas os campos:
- 'Ano e mês do lançamento',
- 'Nome Função' e 
- 'Valor Liquidado'. 

A seguir, serão realizadas as transformações necessárias nos dados brutos.

```python
# Excluir as colunas desnecessárias

selected_columns = ['Ano e mês do lançamento', 'Nome Função','Valor Liquidado (R$)']

df_despesa_exec = df_despesa_exec[selected_columns]

#Demais transformações

transformation_applied = dp.DataTransformation(df_despesa_exec)

#Transformar coluna string em valor numérico

transformation_applied.replace_characters('Valor Liquidado (R$)', ',', '.')

transformation_applied.convert_to_numeric('Valor Liquidado (R$)')

#separando a coluna 'Ano e mês do lançamento' em 2 colunas 'ano' e 'mes'

transformation_applied.split_column('Ano e mês do lançamento', r'/', ['ano', 'mes'])

#Renomeando as colunas para facilitar as análises futuras:

old_column_names = ['Nome Função','Valor Liquidado (R$)']
column_renames = ['despesa_exec', 'valor']

transformation_applied.rename_columns(old_column_names, column_renames)

#agrupando os dados por despesa_exec
                                
display(df_despesa_exec)
```

<font size="6">3. Análise do <u>Orçamento de Despesas</u><font size="5">

```python
df_orcamento_despesa.head(10)
```

De maneira semelhante aos dados relativos a execução das despesas, é necessário escolher os campos relevantes para a análise.

```python
dp.ListFormatter(df_orcamento_despesa.columns).print_formatted()
```

<br>
Diferentemente da Execução da Despesa, o orçamento possui duas colunas de valores relevantes: 'ORÇAMENTO INICIAL' e 'ORÇAMENTO ATUALIZADO'. O que ocorre é que o congresso nacional pode aprovar mudanças no orçamento inicial e, quando isso ocorre, uma entrada no 'ORCAMENTO ATUALIZADO' é realizada. 
<br>
O dataframe será divido em dois, o primeiro contendo os valores iniciais e o segundo os valores atualizados. Além disso, é necessário escolher um nome para a despesa orçada. Para manter a consistência com os dados da despesa executada, a melhor opção é a coluna 'NOME FUNÇÃO.

<font size="5">3.1 Conclusão:</font>

Do conjunto de dados inicial, são necessários para fins deste projeto apenas os campos:
- 'EXERCÍCIO',
- 'NOME FUNÇÃO '
- 'ORÇAMENTO INICIAL (R$)'
- 'ORÇAMENTO ATUALIZADO (R$)' 

A seguir, serão realizadas as transformações necessárias nos dados brutos.

```python
#Dividir o dataframe em inicial e atualizado/ Excluir Colunas desnecessárias

df_orcamento_despesa_ini = df_orcamento_despesa[['EXERCÍCIO', 'NOME FUNÇÃO','ORÇAMENTO INICIAL (R$)']]
df_orcamento_despesa_atu = df_orcamento_despesa[['EXERCÍCIO', 'NOME FUNÇÃO','ORÇAMENTO ATUALIZADO (R$)']]

transformation_applied_ini = dp.DataTransformation(df_orcamento_despesa_ini)
transformation_applied_atu = dp.DataTransformation(df_orcamento_despesa_atu)

#Transformar coluna string em valor numérico

transformation_applied_ini.replace_characters('ORÇAMENTO INICIAL (R$)', ',', '.')
transformation_applied_atu.replace_characters('ORÇAMENTO ATUALIZADO (R$)', ',', '.')

transformation_applied_ini.convert_to_numeric('ORÇAMENTO INICIAL (R$)')
transformation_applied_atu.convert_to_numeric('ORÇAMENTO ATUALIZADO (R$)')

#Renomeando as colunas para facilitar as análises futuras:

old_column_names_ini = ['EXERCÍCIO', 'NOME FUNÇÃO','ORÇAMENTO INICIAL (R$)']
column_renames_ini = ['ano', 'despesa_orc','valor_ini']

old_column_names_atu = ['EXERCÍCIO', 'NOME FUNÇÃO','ORÇAMENTO ATUALIZADO (R$)']
column_renames_atu = ['ano', 'despesa_orc','valor_atu']

transformation_applied_ini.rename_columns(old_column_names_ini, column_renames_ini)
transformation_applied_atu.rename_columns(old_column_names_atu, column_renames_atu)
                                
display(pd.concat([df_orcamento_despesa_ini, df_orcamento_despesa_atu], axis = 1))
```

Repetindo os passos das análises anteriores:

```python
column_names = ['CATEGORIA ECONÔMICA', 'ORIGEM RECEITA', 'ESPÉCIE RECEITA', 'DETALHAMENTO'] #campos a serem analisados

dp.DataFrameProcessor(df_receitas, column_names).process_data() #Mostra, lado a lado, os valores únicos de cada coluna escolhida para a análise.
```

<font size="5">3.1 Conclusão:</font>

Do conjunto de dados inicial, são necessários para fins deste projeto apenas os campos:

- 'DATA LANÇAMENTO' (que será usado para criar os campos 'mes' e 'ano', visando a consistência com as demais tabelas)
- 'ORIGEM RECEITA'
- 'VALOR PREVISTO ATUALIZADO'
- 'VALOR REALIZADO' 

Assim como no dataframe referente ao orçamento, as receitas serão dividos em dois dataframes, o primeiro contendo valores previstos e o segundo, valores realizados. A seguir, serão realizadas as transformações necessárias nos dados brutos.

```python
#Dividir o dataframe em inicial e atualizado/ Excluir Colunas desnecessárias

df_receitas_prev = df_receitas[['DATA LANÇAMENTO', 'ORIGEM RECEITA','VALOR PREVISTO ATUALIZADO']]
df_receitas_real = df_receitas[['DATA LANÇAMENTO', 'ORIGEM RECEITA','VALOR REALIZADO']]

transformation_applied_prev = dp.DataTransformation(df_receitas_prev)
transformation_applied_real = dp.DataTransformation(df_receitas_real)

#Transformar coluna string em valor numérico

transformation_applied_prev.replace_characters('VALOR PREVISTO ATUALIZADO', ',', '.')
transformation_applied_real.replace_characters('VALOR REALIZADO', ',', '.')

transformation_applied_prev.convert_to_numeric('VALOR PREVISTO ATUALIZADO')
transformation_applied_real.convert_to_numeric('VALOR REALIZADO')

#Renomeando as colunas para facilitar as análises futuras:

old_column_names_prev = ['DATA LANÇAMENTO', 'ORIGEM RECEITA','VALOR PREVISTO ATUALIZADO']
column_renames_prev = ['data', 'receita','valor_receita_prev']

old_column_names_real = ['DATA LANÇAMENTO', 'ORIGEM RECEITA','VALOR REALIZADO']
column_renames_real = ['data', 'receita','valor_receita_real']

transformation_applied_prev.rename_columns(old_column_names_prev, column_renames_prev)
transformation_applied_real.rename_columns(old_column_names_real, column_renames_real)

#Criar os campos 'mes' e 'ano'

transformation_applied_prev.split_column('data', r'/', ['dia','ano', 'mes'])
transformation_applied_real.split_column('data', r'/', ['dia','ano', 'mes'])

#agrupando os dados por receita

df_receitas_prev = df_receitas_prev[['receita', 'mes', 'ano', 'valor_receita_prev']]
df_receitas_real = df_receitas_real[['receita', 'mes', 'ano', 'valor_receita_real']]
                                
display(pd.concat([df_receitas_prev, df_receitas_real], axis = 1))
```
